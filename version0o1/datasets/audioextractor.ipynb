```python
import os
from moviepy.video.io.VideoFileClip import VideoFileClip

# Define the input and output folders
sarcastic_folder = "T:/DSV2/S verified"
nonsarcastic_folder = "T:/DSV2/NS verified"
sarcastic_output_folder = "T:/DSV2/S verified audio"
nonsarcastic_output_folder = "T:/DSV2/NS verified audio"

# Ensure the output folders exist
os.makedirs(sarcastic_output_folder, exist_ok=True)
os.makedirs(nonsarcastic_output_folder, exist_ok=True)

def extract_audio(video_path, output_folder):
    """
    Extracts audio from a video and saves it as a .wav file.
    
    Args:
        video_path (str): Path to the video file.
        output_folder (str): Folder to save the extracted audio.
    """
    try:
        # Load the video
        video = VideoFileClip(video_path)
        
        # Generate output file path
        video_name = os.path.splitext(os.path.basename(video_path))[0]
        output_path = os.path.join(output_folder, f"{video_name}.wav")
        
        # Write audio to .wav file
        video.audio.write_audiofile(output_path, verbose=False, logger=None)
        print(f"Audio saved: {output_path}")
        
        video.close()
    except Exception as e:
        print(f"Error processing {video_path}: {e}")

# Process all videos in the sarcastic folder
print("Processing sarcastic videos...")
for file_name in os.listdir(sarcastic_folder):
    if file_name.endswith((".mp4", ".avi", ".mov", ".mkv")):  
        video_path = os.path.join(sarcastic_folder, file_name)
        extract_audio(video_path, sarcastic_output_folder)

# Process all videos in the nonsarcastic folder
print("Processing nonsarcastic videos...")
for file_name in os.listdir(nonsarcastic_folder):
    if file_name.endswith((".mp4", ".avi", ".mov", ".mkv")):  
        video_path = os.path.join(nonsarcastic_folder, file_name)
        extract_audio(video_path, nonsarcastic_output_folder)

print("Audio extraction complete.")
```
